```python
import requests
import pandas as pd

BASE_URL = "https://real-time-amazon-data.p.rapidapi.com/search"

headers = {
    "x-rapidapi-host": "real-time-amazon-data.p.rapidapi.com",
    "x-rapidapi-key": "1dec4b3da8mshb022dd3a6227c9bp1ed4a5jsn43c754acb595"
}

all_products = []  # tempat menyimpan hasil semua page
total_pages_to_fetch = 10 # total page yang ingin diambil

for page in range(1, total_pages_to_fetch + 1):
    params = {
        "query": "Phone",
        "page": page,
        "country": "SG",
        "sort_by": "BEST_SELLERS",
        "product_condition": "ALL",
        "is_prime": "false",
        "deals_and_discounts": "NONE"
    }

    response = requests.get(BASE_URL, headers=headers, params=params)

    print(f"Fetching page {page}: Status {response.status_code}")

    if response.status_code != 200:
        print("Stop karena error!")
        break

    data = response.json()

    # Ambil daftar produk
    products = data["data"]["products"]

    if len(products) == 0:
        print("Tidak ada data lagi, stop.")
        break

    all_products.extend(products)  # gabungkan data ke list besar

# Convert ke DataFrame
df = pd.DataFrame(all_products)
df.head()
```
